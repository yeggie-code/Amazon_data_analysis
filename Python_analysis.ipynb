```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('amazon.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').astype(int)
df.info()
```

```python
df['category'].unique()
```

```python
df['discount_amount'] = df['actual_price'] - df['discounted_price']
df.head()
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '').astype(float)/100
df.head()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['discount_percentage'].corr(df['rating'])
```

```python
df_no_duplicates = df.drop_duplicates(subset="product_id", keep="first")
```

```python
df.to_csv("cleaned_amazon_data.csv", index=False)
```

```python
category_sales = (
    df.groupby('category')['discounted_price'].sum().reset_index().sort_values(by='discounted_price', ascending=False)
)

# Display the top-selling categories
category_sales
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(
    data=df,
    x='discount_percentage',
    y='rating_count',
    scatter_kws={'alpha': 0.5},
    line_kws={'color': 'red'}
)
plt.title(f'Relationship Between Discount Percentage and Rating Count')
plt.xlabel('Discount Percentage (%)')
plt.ylabel('Rating Count')
plt.grid(alpha=0.3)
plt.show()
```

```python
top_rated_products = df.sort_values(by='rating', ascending=False).head(10)
plt.bar(top_rated_products['product_name'], top_rated_products['rating'])
plt.title('Top 10 Highest-Rated Products')
plt.xlabel('Product Name')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()


```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x='discount_percentage',
    y='rating',
)

plt.title('Scatter Plot: Discount Percentage vs. Rating')
plt.xlabel('Discount Percentage')
plt.ylabel('Rating')
plt.xlim(0, 100)  
plt.ylim(0, 5)    
plt.show()
```

```python

df['estimated_sales'] = df['discounted_price'] * df['rating_count']
category_sales = (
    df.groupby('category')['estimated_sales'].sum().reset_index().sort_values(by='estimated_sales', ascending=False)
)
top_categories = category_sales.head(10)
plt.pie(
    top_categories['estimated_sales'],labels=top_categories['category'],autopct='%1.1f%%')

plt.title('Top-Selling Product Categories (Estimated Sales)', fontsize=16)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix)
plt.title('Correlation Matrix')
plt.show()
```
